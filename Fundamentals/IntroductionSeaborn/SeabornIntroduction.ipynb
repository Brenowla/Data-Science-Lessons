```python
dados = pd.read_csv('tips.csv')
```

```python
renomear = {
    'total_bill' : 'valor_da_conta',
    'tip' : 'gorjeta',
    'dessert' : 'sobremesa',
    'day' : 'dia_da_semana',
    'time' : 'hora_do_dia',
    'size' : 'total_de_pessoas'
}
```

```python
gorjetas = dados.rename(columns = renomear)
```

```python
gorjetas.sobremesa.unique()
```

```python
sim_nao ={
    'No': 'Não',
    'Yes': 'Sim'
}
```

```python
gorjetas.sobremesa = gorjetas.sobremesa.map(sim_nao)
```

```python
gorjetas
```

```python
gorjetas.dia_da_semana.unique()
```

```python
dias_semana = {
    'Sun': 'Domingo', 
    'Sat': 'Sábado',
    'Thur': 'Quinta',
    'Fri': 'Sexta'
}
```

```python
gorjetas.dia_da_semana = gorjetas.dia_da_semana.map(dias_semana)
```

```python
gorjetas.hora_do_dia.unique()
```

```python
hora = {
    'Dinner': 'Jantar',
    'Lunch': 'Almoço'
}
```

```python
gorjetas.hora_do_dia = gorjetas.hora_do_dia.map(hora)
```

## Teste de hpótese

### Hipotese nula: A distribuição da gorjeta é a mesma nos dois grupos

### Hipotese alternativa: : A distribuição da gorjeta é a diferente nos dois grupos

```python
from scipy.stats import ranksums
```

```python
sobremesa = gorjetas.query("sobremesa == 'Sim'").porcentagem
```

```python
sem_sobremesa = gorjetas.query("sobremesa == 'Não'").porcentagem
```

```python
ranksums(sobremesa, sem_sobremesa)
```
