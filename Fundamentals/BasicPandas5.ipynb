## Seleções e Frequências

```python
import pandas as pd
```

```python
dados = pd.read_csv('data/aluguel_residencial.csv', sep = ';')
```

```python
dados.head()
```

```python
# Selecione imóveis classificados com "Apartamento"
selecao = dados['Tipo'] == 'Apartamento'
n1 = dados[selecao]
n1
```

```python
n1.shape[0]
```

```python
# Selecione os imóveis classificados com tipo 'Casa', 'Casa de Condomínio' e 'Casa de Vila'
selecao =( dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Condomínio') | (dados['Tipo'] == 'Casa de Vila')
n2 = dados[selecao]
n2
```

```python
n2.shape[0]
```

```python
# Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites
selecao = (dados['Area'] >=60) & (dados['Area'] <= 100)
n3 = dados[selecao]
n3
```

```python
n3.shape[0]
```

```python
# Selecione os ímoveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00
selecao = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000)
n4 = dados[selecao]
n4
```

```python
n4.shape[0]
```
